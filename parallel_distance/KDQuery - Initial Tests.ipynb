```python
x = array([   10,   100,  1000, 10000])
kdq_mac_par = array([ 0.01303315,  0.06504107,  0.40919018,  4.77737594])
kdq_mac_ser = array([0.00257301330566,0.0680389404297, 0.704342842102, 8.97821187973 ])
kdq_win_par = array([1.829999999, 0.742, 1.857, 3.8659])
kdq_win_ser = array([0.0019, 0.024999, 0.363999, 4.89900])
```

```python
plot(x, kdq_mac_par,'r--', label="KDQuery Parallel", linewidth=2)
plot(x, kdq_mac_ser,'r', label="KDQuery Serial", linewidth=2)
figsize(12,8)
title("OSX KDQuery Benchmark Tests")
legend(loc=4)
xlabel("$n$")
ylabel("Time (s)")
grid()
```

```python
plot(x, kdq_win_par,'b--', label="KDQuery Parallel", linewidth=2)
plot(x, kdq_win_ser, label="KDQuery Serial", linewidth=2)
figsize(12,8)
title("Windows X KDQuery Benchmark Tests")
legend(loc=4)
xlabel("$n$")
ylabel("Time (s)")
grid()
```

```python
plot(x, kdq_mac_par,'r', label="KDQuery OS X Parallel", linewidth=2)
plot(x, kdq_win_par, label="KDQuery WIN Parallel", linewidth=2)
figsize(12,8)
title("OSX v. Win KDQuery Benchmark Tests")
legend(loc=4)
xlabel("$n$")
ylabel("Time (s)")
grid()
```
